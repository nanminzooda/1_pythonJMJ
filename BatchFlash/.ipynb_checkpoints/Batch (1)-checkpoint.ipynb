```python
from math import *
from sympy import *
from scipy.optimize import fsolve

x, y, z, t = symbols('x y z t')
f, g, h = symbols('f, g, h', cls=Function)

init_printing()
%matplotlib inline

print("yD(최종 도수의 몰분율)값과 반복 횟수를 입력해주세요.")
print("yD: ")
yD = float(input()) # 설정한 소주의 도수
print("반복 횟수: ")
BN = int(input()) # Batch Numbers
print("F/B의 최솟값을 입력해주세요.")
kmin = float(input())
print("F/B의 최댓값을 입력해주세요.")
kmax = float(input())

def f(x): # Rayleigh equation
    return ((1-xB)**a)*x - (k**(a-1))*((1-x)**a)*xB

print('   F/B    BN    T[C]       a         xF         xB         yD      Q[MJ]     Q_tot[MJ]     F[L]    Steam needed[kg]')

HF = -1667.4 # H-xy diagram의 reference가 78C. feed 온도를 25C로 가정. 조성에 따라 HF 변화 없음.

D_L = 80 #L/hr

Me = 46.07 # g/mol, 에탄올 분자량
Mw = 18.01528 # g/mol, 물 분자량

h = 2162600 # J/g, 열전달계수 of steam at 2 barg(대기압과의 상대압력)

target = yD

# product 분자량, 밀도
MP = Me*target + Mw*(1-target)
w_p = Me*target*100/(Me*target + Mw*(1-target)) # product의 질량%

if 0 < w_p < 10: # 0~10%, feed의 밀도 구하기
    den_p = (0.980*w_p + 0.997*(10-w_p))/10
elif 10 <= w_p < 20: # 10~20%
    den_p = (0.966*(w_p-10) + 0.980*(20-w_p))/10
elif 20 <= w_p < 30: # 20~30%
    den_p = (0.951*(w_p-20) + 0.966*(30-w_p))/10
elif 30 <= w_p < 40: # 30~40%
    den_p = (0.931*(w_p-30) + 0.951*(40-w_p))/10
else: # 40~50%
    den_p = (0.910*(w_p-40) + 0.931*(50-w_p))/10

# D L/hr -> mol/hr
D_mol = D_L*1000*den_p/MP

k = kmin
while(k<=kmax):
    i = 0
    yD = target
    Q_tot = 0
    D = D_mol # 초기 D값
    
    while(i<BN): # batch number만큼 진행(뒤에서부터)
        # 온도 구하기
        T_ydia = -151.61*x**5 + 351.27*x**4 - 238.11*x**3 + 47.14*x**2 - 30.951*x + 100.5
        T_xdia = 1279.9*x**6 - 4393*x**5 + 5914.3*x**4 - 3948.5*x**3 + 1367.9*x**2 - 242.11*x + 99.91
        T = float(T_ydia.subs(x, yD)) # C
        xB = float(solve(T_xdia-T)[0])

        # T -> a 구하기
        Pe = 10**(7.68117 - 1332.04/(T+199.2)) # DDB data / mmHg / 70C ~ 243C
        Pw = 10**(8.07131 - 1730.63/(T+233.426)) # 1C~100C
        a = Pe/Pw
        
        # 뒤에서 i+1번째 횟수의 Batch의 feed 구하기(F)
        xF = float(fsolve(f, 0.1))

        
        # F 분자량 구하기
        MF = Me*xF + Mw*(1-xF) # g/mol, 피드의 분자량
        
        # F 밀도 구하기
        w = Me*xF*100/(Me*xF + Mw*(1-xF)) # Feed의 질량%
        if 0 < w < 10: # 0~10%, feed의 밀도 구하기
            den = (0.980*w + 0.997*(10-w))/10
        elif 10 <= w < 20: # 10~20%
            den = (0.966*(w-10) + 0.980*(20-w))/10
        elif 20 <= w < 30: # 20~30%
            den = (0.951*(w-20) + 0.966*(30-w))/10
        elif 30 <= w < 40: # 30~40%
            den = (0.931*(w-30) + 0.951*(40-w))/10
        else: # 40~50%
            den = (0.910*(w-40) + 0.931*(50-w))/10
        
        # H 이용해서 Q 구하기
        H_xdia = -6e-8*x**5 + 2e-7*x**4 - 2e-7*x**3 + 6e-8*x**2 + 805.17*x + 1632.9 # J/mol
        H_ydia = 3102.9*x**5 - 6803.2*x**4 + 12332*x**3 - 13245*x**2 + 1331.1*x + 41836 # J/mol
        H_x = H_xdia.subs(x, xB) # H-x diagram에 xB를 넣었을 때의 H값
        H_y = H_ydia.subs(x, yD) # H-y diagram에 yD를 넣었을 때의 H값
        
        F_L = D_L/(1-1/k)**(i+1) # L/hr
        F = F_L/MF*den*1000 # mol/hr
        D = F # 뒤에서 i+2 번째 횟수로 이동
        
        Q = float(((1-1/k)*H_y+1/k*H_x-HF)*F) # J/hr
        Q_hr = Q*0.000001 # MJ/hr
        Q_tot = Q_tot + Q_hr # MJ/hr

        # h(steam의 열전달계수)를 이용하여 ms(steam 양) 구하기
        ms = Q/h*0.001 # kg

        print('   %2.2f  %2d    %3.2f  %9.6f  %9.6f  %9.6f  %9.6f %9.2f  %9.2f  %9.2f  %9.6f' %(k, BN-i, T, a, xF, xB, yD, Q_hr, Q_tot, F_L, ms))
        # 다음 Batch로 이동하기 위해 F를 D로 변경
        yD = xF
        if yD <= 0.01937: break
        i = i + 1

    k = k + 0.01
```
