```python
from math import *
from sympy import *
from scipy.optimize import fsolve

x, y, z, t = symbols('x y z t')
f, g, h = symbols('f, g, h', cls=Function)

init_printing()
%matplotlib inline

print('yD값과 단수를 입력해주세요.')
print('yD: ')
target = float(input()) # mole fraction of target
#target = 0.07146
print('단수: ')
FN = int(input()) # Flash Number
#FN = int(6)
print('f의 최솟값: ')
f_min = float(input()) # D/F
print('f의 최댓값: ')
f_max = float(input())

HF = -1667.4 # H-xy diagram의 reference가 78C. feed 온도를 25C로 가정. 조성에 따라 HF 변화 없음.

D_L = 80 #L/hr

Me = 46.07 # g/mol, 에탄올 분자량
Mw = 18.01528 # g/mol, 물 분자량

f = f_min

print('   f   FN      xB        yD       T[C]       Q[MJ/hr]   Q_tot[MJ/hr]  F[L/hr]')

# product 분자량, 밀도
MP = Me*target + Mw*(1-target)
w_p = Me*target*100/(Me*target + Mw*(1-target)) # product의 질량%

if 0 < w_p < 10: # 0~10%, feed의 밀도 구하기
    den_p = (0.980*w_p + 0.997*(10-w_p))/10
elif 10 <= w_p < 20: # 10~20%
    den_p = (0.966*(w_p-10) + 0.980*(20-w_p))/10
elif 20 <= w_p < 30: # 20~30%
    den_p = (0.951*(w_p-20) + 0.966*(30-w_p))/10
elif 30 <= w_p < 40: # 30~40%
    den_p = (0.931*(w_p-30) + 0.951*(40-w_p))/10
else: # 40~50%
    den_p = (0.910*(w_p-40) + 0.931*(50-w_p))/10

# D L/hr -> mol/hr
D_mol = D_L*1000*den_p/MP

while(f<=f_max): # 목표에 도달할 때까지
    i = 0
    xF = 0.0193 # mole fraction of feed
    Q_tot = 0
    D = D_mol # 최초값 설정
    
    for i in range(0,FN): # 단수
        ol = (f-1)/f * ( x - xF) + xF  # operating line
        y_xdia = 8.6693*x-53.57103*x**2+162.90856*x**3-251.69406*x**4+191.23992*x**5-56.55869*x**6
        xB = solve(y_xdia-ol,x)[0] # ol과 y-x diagram 교점의 x좌표
        yD = y_xdia.subs(x, xB) # ol과 y-x diagram 교점의 y좌표
        
        H_xdia = -6e-8*x**5 + 2e-7*x**4 - 2e-7*x**3 + 6e-8*x**2 + 805.17*x + 1632.9
        H_ydia = 3102.9*x**5 - 6803.2*x**4 + 12332*x**3 - 13245*x**2 + 1331.1*x + 41836
        H_x = H_xdia.subs(x, xB) # H-x diagram에 xB를 넣었을 때의 H값
        H_y = H_ydia.subs(x, yD) # H-y diagram에 yD를 넣었을 때의 H값
        
        # 분자량 구하기
        MF = Me*xF + Mw*(1-xF) # g/mol, 피드의 분자량
        
        # 밀도 구하기
        w = Me*xF*100/(Me*xF + Mw*(1-xF)) # Feed의 질량%
        if 0 < w < 10: # 0~10%, feed의 밀도 구하기
            den = (0.980*w + 0.997*(10-w))/10
        elif 10 <= w < 20: # 10~20%
            den = (0.966*(w-10) + 0.980*(20-w))/10
        elif 20 <= w < 30: # 20~30%
            den = (0.951*(w-20) + 0.966*(30-w))/10
        elif 30 <= w < 40: # 30~40%
            den = (0.931*(w-30) + 0.951*(40-w))/10
        else: # 40~50%
            den = (0.910*(w-40) + 0.931*(50-w))/10
        
        # 온도 구하기
        T_ydia = -151.61*x**5 + 351.27*x**4 - 238.11*x**3 + 47.14*x**2 - 30.951*x + 100.5
        T_xdia = 1279.9*x**6 - 4393*x**5 + 5914.3*x**4 - 3948.5*x**3 + 1367.9*x**2 - 242.11*x + 99.91
        T_x = T_xdia.subs(x, xB)
        T_y = T_ydia.subs(x, yD)
        T = (T_x + T_y)/2 # C

        # Q 구하기
        H_F_x = f*yD + (1-f)*xB # H-xy diagram에서 F의 x값, J/mol
        H_F_y = f*H_y + (1-f)*H_x # H-xy diagramd에서 F의 y값, J/mol
        Q_prime = H_F_y
        
        F_L = D_L/(f**(FN-i)) # L/hr
        F = F_L/MF*den*1000 # mol/hr
        D = F
        
        Q = (Q_prime-HF)*F # J/hr
        Q_hr = Q*0.000001 # MJ/hr
        Q_tot = Q_tot + Q_hr # MJ/hr
        print('%.4f  %d  %9.6f  %9.6f  %3.2f  %11.2f  %11.2f     %5.2f' %(f, i+1, xB, yD, T, Q_hr, Q_tot, F_L))
        if yD>=target: break # target mole fraction에 도달하면 stop
        xF = yD # 다음 단으로 이동하기 위해 교점의 y좌표를 feed mole fraction으로 지정

    f += 0.0001

    
```
